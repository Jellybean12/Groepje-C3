#### Maak gebruik van regular expressions om een lijst te krijgen met alle woorden in de tekst.

## Data preparation

```python
# Opbouwen van een woordenlijst die bestaat uit alle woorden in de tekst zonder hoofdletters
# maak een functie 'woorden' met als input 'InputTekst' die alle woorden in een tekst in een vector 'woordenlijst' zet


def woorden (InputTekst):
    woordenlijst = re.findall(r"['\w']+", InputTekst.lower())
    return woordenlijst
```

Opbouwen van een vector van woorden

```python
woordenlijst = woorden(tekst)
```

```python
# Bekijk de opgebouwde woordvector
woordenlijst
```

Je wilt de woordenlijst opschonen en de zogeheten stopwoorden hier uit verwijderen.
De module corpus van nltk bevat een package met stopwoorden in verschillende talen. Wij gebruiken nu de Nederlandse stopwoorden om onze tekst op te schonen.

```python
from nltk.corpus import stopwords
stopwoorden = stopwords.words("dutch")
```

```python
print(stopwoorden)
```

```python
# Maak een functie die de stopwoorden uit een tekst haalt, waarbij je kan aangeven welke taal de tekst heeft.


def StopwoordenVerwijderen (input, language):
    from nltk.corpus import stopwords
    stopwoorden = stopwords.words(language)
    schoon = [woord for woord in input if woord not in stopwoorden]
    #maak een var woord aan. Zolang woord in input lees volgende. Als niet in stopwoorden stop dan in schoon
    return schoon
```

```python
# roep de functie 'StopwoordenVerwijderen' aan met de juiste parameters en stop het resultaat in de gegeven variabele

woordenlijst_opgeschoond = StopwoordenVerwijderen(woordenlijst, "dutch")
```

#### Het is nu makkelijk om te bepalen hoeveel woorden er uit de tekst verwijderd zijn.

```python
print ("Er zijn ", len(woordenlijst)-len(woordenlijst_opgeschoond)," stopwoorden verwijderd uit de tekst.") 
```

### Vervolgens gaan we een bag of words opbouwen (welke woorden komen er voor in de tekst en hoevaak?).

```python
#origineel
def tellen(vector):
    freq={}
    for woord in vector:
        if woord in freq:
            freq[woord] +=1
        else:
            freq[woord] = 1
    sorted_freq= sorted(freq.items(), key=operator.itemgetter(1), reverse=True)
    return sorted_freq
```

```python
BagofWords = tellen(woordenlijst_opgeschoond) 
```

```python
BagofWords
```

Maak een grafiek die weergeeft hoevaak een woord wordt gebruikt in een tekst. Op deze manier kun je in √©√©n oogopslag belangrijke woorden vaststellen.

```python
bow = pd.DataFrame(BagofWords, columns=['woord','aantal'])
```

```python
bow
```

```python
bow.plot(x='woord', kind='bar', figsize=(10,10))
```

Van de bovenstaande stappen maken we ook weer een functie zodat het hergebruik vereenvoudigd wordt.

```python
#Groepeer de functies de je hierboven hebt gebruikt in √©√©n Jupyter Notebook cel. Herschrijf functies waar nodig.
#Definieer een functie bowChart die een histogram geeft waaruit je kunt aflezen hoevaak de diverse woorden voorkomen.
#Defineer een functie bowTable die een DataFrame uitprint waaruit je kunt aflezen hoevaak de diverse woorden voorkomen.


def bowTable(chartInput, language):
    # Interpunctie verwijderen
    woordenlijst = re.findall(r"['\w']+", chartInput)
        
    # Remove stopwoorden
    from nltk.corpus import stopwords
    stopwoorden = stopwords.words(language)
    schoon = [woord for woord in woordenlijst if woord.lower() not in stopwoorden]
    
    # Count
    freq={}
    for woord in schoon:
        if woord in freq:
            freq[woord] +=1
        else:
            freq[woord] = 1
    sorted_freq= sorted(freq.items(), key=operator.itemgetter(1), reverse=True)
    
    # Add to DataFrame
    table = pd.DataFrame(sorted_freq, columns=['woord','aantal'])
    return table

def bowChart(tableInput, language):
    # Get dataframe
    table = bowTable(tableInput, language)
    
    # Return plot
    return table.plot(x='woord', kind='bar', figsize=(20,10))
```

```python
bowTable(tekst,'Dutch')
```

```python
bowChart(tekst,'Dutch')
```
